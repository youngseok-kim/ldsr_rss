## Extension of LD score regression

In this documentation, we study the LD score regression with the WTCCC height GWAS data. First we load the data and take only data from Chromosome 1.

```julia
using PyPlot, MAT
```

```julia
data = matread("height.mat");
```

```julia
X = convert(Array{Float64,2},data["X"]);
ind = find(data["chr"] .== 1);
X = X[:,ind];
n, p = size(X);
```

Now we center and normalize each column to have mean 0 and variance 1.

```julia
X = X - repmat(mean(X,1),n,1);
X = X ./ sqrt.(sum(X.^2,1)) * sqrt(n);
```

```julia
temp = X[:,1000]'*X/n;

### Histogram
ax = axes();
h = plt[:hist](temp,200); # Histogram
grid("on");
xlabel("X");
ylabel("Y");
title("Histogram");
```

```julia
R_hat = sparse(Diagonal(ones(p))); d = 5;
for i = 1:p-d
    R_hat[(i+1):(i+d),i] = X[:,(i+1):(i+d)]' * X[:,i]/n;
    R_hat[i,(i+1):(i+d)] = R_hat[(i+1):(i+d),i]';
end
```

```julia
## LD score
ld1 = sum(temp.^2) - p/n

## Fast calculation of LD score
ld2 = sum(R_hat[:,m].^2)

##
[ld1 ld2]
```

Without any confounder, suppose $X_1,\cdots,X_n$ are i.i.d. $N(0,R_{p\times p})$. Note that we have assumed the normality. Consider $y|X,\beta \sim N(X\beta,\sigma_e^2 I)$ where $X \in\mathbb{R}^{n\times p}$, $\beta \in \mathbb{R}^p$. We further assume that $\beta \sim N(0,(\sigma_g^2/p)I_p)$. More precisely, in terms of the matrix normal distribution, $X \sim MN(0,I_{n \times n},R_{p \times p})$. $R$ is a correlation matrix. At the population level, if $y \sim N(0,\Sigma)$ then
\begin{equation}
X^T \Sigma X = (\sigma_g^2 /p) (X^T X)^2 + \sigma_e^2 (X^T X)
\end{equation}
or equivalently
\begin{equation}
\hat\beta \hat\beta^T = (\sigma_g^2/p) \hat{R}^2 + n^{-1}\sigma_e^2 \hat{R}
\end{equation}
where $\hat\beta = n^{-1} X^T y$ and $\hat{R} = n^{-1} X^T X$. This yields
\begin{equation}
n \hat\beta_j^2 \approx (n\sigma_g^2/p)\hat{l}_j + \sigma_e^2
\end{equation}
Also, the estimation of LD score would be
\begin{equation}
\hat{l}_j = \sum_{k=1}^p \hat{r}_{jk}^2 \approx l_j + (p/n)
\end{equation}
where the bias $p/n$ is from the high-dimensionality.

```julia
n = 1000; p = 10000;
Sigma = SymTridiagonal(ones(p),ones(p-1)/2);
X = randn(n,p) * sqrtm(full(Sigma));
X = X./sqrt.(sum(X.^2,1)) * sqrt(n);
```

```julia
R_hat = X'*X/n;
l_hat = sum(R_hat.^2,2);
plot(l_hat)
```

```julia
pcolormesh(R_hat[1:100,1:100])
```

```julia
mean(l_hat)
```

Now we sample $\beta$ and $y$ as follows.

```julia
beta = randn(p) / sqrt(p) * 4;
y = X * beta + randn(n);
beta_hat = X'*y/n;
plot(beta)
```

```julia
plot(l_hat,n*beta_hat.^2)
```

```julia
temp = [ones(p) l_hat - p/n];
reg = (temp'*temp)\(temp'* n*beta_hat.^2)
```

Let's compare this with the theory. Note that the slope is $n\sigma_g^2/p = 1.6$ and the intercept is $\sigma_g^2 + \sigma_e^2 = 17$

```julia
b = 100;
l_min = minimum(l_hat); l_max = maximum(l_hat);
diff = l_max-l_min;
l_bin = zeros(b); n_beta_hat_bin = zeros(b);
for i = 1:b
    ind = find( (l_hat .<= l_min + i * diff/b) .& (l_hat .>= l_min + (i-1) * diff/b) );
    l_bin[i] = l_min + (i+0.5) * diff/b;
    n_beta_hat_bin[i] = mean(n*beta_hat[ind].^2);
end
```

```julia
plot(l_bin - p/n,n_beta_hat_bin)
```

```julia
ind = find(.~isnan.(n_beta_hat_bin));
temp = [ones(length(ind)) l_bin[ind] - p/n];
reg = (temp'*temp)\(temp'*n_beta_hat_bin[ind])
```

Let's compare this with the theory. It must be the case that $n\hat\beta_j^2 \approx (n\sigma_g^2/p) \hat{l}_j + \sigma_e^2 \approx (n\sigma_g^2/p)l_j + (\sigma_g^2 + \sigma_e^2)$

Now let's assume that there is a confounding effect. For simplicity, let $X_j = f Z_j + \tilde{X}_j$ where $Z_j \in \{-1,1\}$ a confounding factor, $f \in N(0,F_{ST}V)$ a drift term or a loading, and $\tilde{X}$ is a confounding adjusted design matrix, i.e. $\tilde{X} \sim N(0,I_n,R)$. To model population stratification, we let $y_j = X_j^T\beta + gZ_j + \epsilon_j$ where $g = \sigma_s^2$. Then
\begin{equation}
(z,x,y)|f,g,\beta \sim N\left( \left[\begin{array}{c} 0 \\ 0 \\ 0 \end{array}\right],
\left[ \begin{array}{ccc} 1 & f^T & f^T\beta + g \\
f & R + ff^T & (R + ff^T)\beta + fg \\
f^T\beta + g & \beta^T (R + ff^T) + gf^T & \beta^T (R+ff^T) \beta + 2gf^T\beta + g^2 \end{array} \right]
\right)
\end{equation}
The conditional distribution satisfies
\begin{equation}
(x,y)|z \sim N\left( \left[\begin{array}{c} f z \\ (f^T\beta +g) z \end{array}\right] ,
\left[ \begin{array}{cc} R &  R\beta \\
\beta^T R & \beta^T R \beta
\end{array} \right]
\right)
\end{equation}
